```python
country_codes = df.columns.str.extract(r'\[([A-Z]+)\]', expand=False).unique()

country_data_dict = {}

for code in country_codes:
    country_columns = [col for col in df.columns if f'[{code}]' in col]
    
    country_data = df[['Time'] + country_columns]
    
    country_data.columns = ['Time'] + [col.split('[')[0].strip() for col in country_columns]
    
    country_data_dict[code] = country_data
```

```python
for country, data in country_data_dict.items():
    print(data)
    country_name = data.columns[1].split('[')[0].strip()
    print(country_name)
    
    data['Country'] = country_name
```
